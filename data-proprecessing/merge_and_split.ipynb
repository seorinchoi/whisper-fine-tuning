```python
import os
from glob import glob

# ------------ 경로 설정 ------------
BASE_DIR = "/workspace"
DATA_DIR = os.path.join(BASE_DIR, "data")
LABEL_DIR = os.path.join(BASE_DIR, "label")

# ------------ 재귀적으로 모든 파일 탐색 ------------
wav_files = glob(os.path.join(DATA_DIR, "**", "*.wav"), recursive=True)
json_files = glob(os.path.join(LABEL_DIR, "**", "*.json"), recursive=True)

# ------------ 확장자 제외하고 파일명만 ------------
wav_names = [os.path.splitext(os.path.basename(f))[0] for f in wav_files]
json_names = [os.path.splitext(os.path.basename(f))[0] for f in json_files]

# ------------ 매칭되는 파일과 매칭 안되는 파일 ------------
matched = set(wav_names) & set(json_names)
wav_only = set(wav_names) - set(json_names)
json_only = set(json_names) - set(wav_names)

# ------------ 결과 출력 ------------
print(f"총 WAV 파일: {len(wav_names)}")
print(f"총 JSON 파일: {len(json_names)}")
print(f"매칭되는 파일 수: {len(matched)}")
print(f"WAV만 있는 파일 수: {len(wav_only)}")
print(f"JSON만 있는 파일 수: {len(json_only)}")

# 예시로 처음 10개 확인
print("\nWAV만 있는 파일 예시:", list(wav_only)[:10])
print("JSON만 있는 파일 예시:", list(json_only)[:10])
```

```python
import os
import shutil
from glob import glob

BASE_DIR = "/workspace"
DATA_DIR = os.path.join(BASE_DIR, "data")
LABEL_DIR = os.path.join(BASE_DIR, "label")

def flatten_folder(root_dir, extension):
    """하위 폴더까지 포함해 파일을 최상위 폴더로 이동"""
    files = glob(os.path.join(root_dir, "**", f"*{extension}"), recursive=True)
    print(f"발견된 {extension} 파일 수: {len(files)}")

    for file_path in files:
        file_name = os.path.basename(file_path)
        dest_path = os.path.join(root_dir, file_name)

        # 중복 파일명 처리
        if os.path.exists(dest_path):
            base, ext = os.path.splitext(file_name)
            i = 1
            while os.path.exists(os.path.join(root_dir, f"{base}_{i}{ext}")):
                i += 1
            dest_path = os.path.join(root_dir, f"{base}_{i}{ext}")

        shutil.move(file_path, dest_path)

    # 빈 폴더 삭제
    for dirpath, dirnames, filenames in os.walk(root_dir, topdown=False):
        if not dirnames and not filenames:
            os.rmdir(dirpath)

# ------------ data 폴더 평탄화 ------------
flatten_folder(DATA_DIR, ".wav")
print("✅ data 폴더 평탄화 완료")

# ------------ label 폴더 평탄화 ------------
flatten_folder(LABEL_DIR, ".json")
print("✅ label 폴더 평탄화 완료")
```

```python
# train과 test 폴더 생성 후, 파일명을 기준으로 1:1 매칭
# 랜덤하게 섞은 뒤 8:2로 나눔 

import os
import shutil
import random
from glob import glob

# ------------ 경로 설정 ------------
BASE_DIR = "/workspace"
DATA_DIR = os.path.join(BASE_DIR, "data")
LABEL_DIR = os.path.join(BASE_DIR, "label")

# ------------ train/test 폴더 생성 ------------
TRAIN_DATA_DIR = os.path.join(BASE_DIR, "train", "data")
TRAIN_LABEL_DIR = os.path.join(BASE_DIR, "train", "label")
TEST_DATA_DIR = os.path.join(BASE_DIR, "test", "data")
TEST_LABEL_DIR = os.path.join(BASE_DIR, "test", "label")

for d in [TRAIN_DATA_DIR, TRAIN_LABEL_DIR, TEST_DATA_DIR, TEST_LABEL_DIR]:
    os.makedirs(d, exist_ok=True)

# ------------ 파일 리스트 가져오기 ------------
wav_files = glob(os.path.join(DATA_DIR, "*.wav"))
json_files = glob(os.path.join(LABEL_DIR, "*.json"))

# 확장자 제외한 이름 기준으로 매칭
wav_dict = {os.path.splitext(os.path.basename(f))[0]: f for f in wav_files}
json_dict = {os.path.splitext(os.path.basename(f))[0]: f for f in json_files}

# 매칭되는 파일명만 사용
matched_names = list(set(wav_dict.keys()) & set(json_dict.keys()))
print(f"총 매칭되는 파일 수: {len(matched_names)}")

# ------------ train/test 분할 ------------
random.shuffle(matched_names)
split_idx = int(len(matched_names) * 0.8)
train_names = matched_names[:split_idx]
test_names = matched_names[split_idx:]

# ------------ 파일 이동 함수 ------------
def move_files(names, src_data_dict, src_label_dict, dest_data_dir, dest_label_dir):
    from tqdm import tqdm
    for name in tqdm(names, desc="파일 이동 진행중"):
        shutil.copy2(src_data_dict[name], os.path.join(dest_data_dir, os.path.basename(src_data_dict[name])))
        shutil.copy2(src_label_dict[name], os.path.join(dest_label_dir, os.path.basename(src_label_dict[name])))

# ------------ train 이동 ------------
move_files(train_names, wav_dict, json_dict, TRAIN_DATA_DIR, TRAIN_LABEL_DIR)
# ------------ test 이동 ------------
move_files(test_names, wav_dict, json_dict, TEST_DATA_DIR, TEST_LABEL_DIR)

print("✅ train/test 분할 완료!")
print(f"Train: {len(train_names)} pairs, Test: {len(test_names)} pairs")
```
